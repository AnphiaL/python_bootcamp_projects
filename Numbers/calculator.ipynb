# Calculator 

A simple calculator to do basic operators. Make it a scientific calculator for added complexity.

```python
from tkinter import*
import math
```

```python
root = Tk()
root.title("Jupyter Notebbok Calculator")
root.resizable(width = False, height = False)
root.geometry("510x636+500+40")#adjust the size of the calculator
#looks appropriate in macOS Big Sur 11.6

mainFrame = Frame(root, bd = 20, bg = "gainsboro", relief = RIDGE)
mainFrame.grid()

innerFrame = Frame(mainFrame, bd = 10, bg = "gainsboro", relief = RIDGE)
innerFrame.grid()

class Calculator:
    def __init__(self):
        self.total = 0
        self.current = ""
        self.firstnum = ""
        self.secondnum = ""
        self.input_value = True
        self.check_sum = False
        self.op = ""
        self.result = False
        
    def number_enter(self, num):
        self.result = False
        firstnum = textDisplay.get()
        secondnum = str(num)
        if self.input_value:
            self.current = secondnum
            self.input_value = False
        else:
            if secondnum == ".":
                if secondnum in firstnum:
                    return
            self.current = firstnum + secondnum
        self.display(self.current)
        #added to numberpad command
        
    def sum_of_total(self):
        self.result = True
        self.current = float(self.current)
        if self.check_sum == True:
            self.valid_function()
        else:
            self.total = float(textDisplay.get())
    
    def display(self, value):
        textDisplay.delete(0, END)
        textDisplay.insert(0, value)
    
    def clear_entry(self):
        self.result = False 
        self.current = "0"
        self.input_value = True
        self.display(0)
        #added to buttonClear command
    
    def all_clear(self):
        self.clear_entry()
        self.total = "0"
        #added to buttonAllClear Command
    
    def mathPM(self):
        self.result = False
        self.current = -(float(textDisplay.get()))
        self.display(self.current)
        #added to buttonPM command
    
    def pi(self):
        self.result = False
        self.current = math.pi
        self.display(self.current)
        #added to buttonPi command
        
    def sqrt(self):
        self.result = False
        self.current = math.sqrt(float(textDisplay.get()))
        self.display(self.current)
    
    def one_in(self):
        self.result = False
        self.firstnum = (float(1))
        self.secondnum = (float(textDisplay.get()))
        self.current = (self.firstnum/self.secondnum)
        self.display(self.current)
    
    def squared2(self):
        self.result = False
        self.current = (float(textDisplay.get()))**2
        self.display(self.current)
        
    def squared3(self):
        self.result = False
        self.current = (float(textDisplay.get()))**3
        self.display(self.current)
    
    def sin(self):
        self.result = False
        self.current = math.sin(float(textDisplay.get()))
        self.display(self.current)
    
    def cos(self):
        self.result = False
        self.current = math.cos(float(textDisplay.get()))
        self.display(self.current)
    
    def tan(self):
        self.result = False
        self.current = math.tan(float(textDisplay.get()))
        self.display(self.current)
    
    def euler_number(self):
        self.result = False
        self.current = math.e
        self.display(self.current)
    
    def backspace(self):
        numLen = len(textDisplay.get())
        textDisplay.delete(numLen-1, "end")
        if numLen == -1:
            textDisplay.insert(0, "0")
            
    def valid_function(self):
        if self.op == "add":
            self.total += self.current
        if self.op == "sub":
            self.total -= self.current
        if self.op == "multi":
            self.total *= self.current
        if self.op == "divide":
            self.total /= self.current
        if self.op == "mod":
            self.total %= self.current
        if self.op == "floor_div":
            self.total = self.total//self.current
        if self.op == "exp":
            self.total = self.total**self.current 
        self.input_value = True
        self.check_sum = False
        self.display(self.total)
    
    def operation(self,op):
        self.current = float(self.current)
        if self.check_sum:
            self.valid_function()
        elif not self.result:
            self.total = self.current
            self.input_value = True
        self.check_sum = True
        self.op = op
        self.result = False

added_value = Calculator()

textDisplay = Entry(innerFrame, font = ("arial",18,"bold"), bd = 10, width = 38, justify = RIGHT)
textDisplay.grid(row = 0, column = 0, columnspan = 4, pady = 1)
textDisplay.insert(0, "0")

#create number pad
numberpad = "789456123"
i = 0
btn = []
for j in range(3, 6):
    for k in range(3):
        btn.append(Button(innerFrame, width = 6, height = 2, font = ("arial",18,"bold"), bd = 7, text = numberpad[i]))
        btn[i].grid(row = j, column = k, pady = 1)
        btn[i]["command"] = lambda x =numberpad[i]:added_value.number_enter(x)
        i += 1 #run this and number pod will be ready
        #===========================Button Widgets===========================
        
btnDelete = Button(innerFrame, text = "Delete", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.backspace)
btnDelete.grid(row = 1, column = 0, pady = 1)

btnClear = Button(innerFrame, text = "Clear", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.clear_entry)
btnClear.grid(row = 1, column = 1, pady = 1)

btnAllClear = Button(innerFrame, text = "ClearAll", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.all_clear)
btnAllClear.grid(row = 1, column = 2, pady = 1)

btnPM = Button(innerFrame, text = chr(177), width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.mathPM)
btnPM.grid(row = 1, column = 3, pady = 1)

btnsqrt = Button(innerFrame, text = "√", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, command = added_value.sqrt)
btnsqrt.grid(row = 2, column = 0, pady = 1)

btnSquared2 = Button(innerFrame, text = "X²", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, command = added_value.squared2)
btnSquared2.grid(row = 2, column = 1, pady = 1)

btnSquared3 = Button(innerFrame, text = "X³", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, command = added_value.squared3)
btnSquared3.grid(row = 2, column = 2, pady = 1)

btnPi = Button(innerFrame, text = "π", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.pi)
btnPi.grid(row = 2, column = 3, pady = 1)
        
btnAdd = Button(innerFrame, text = "+", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.operation("add"))
btnAdd.grid(row = 3, column = 3, pady = 1)

btnSub = Button(innerFrame, text = "-", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.operation("sub"))
btnSub.grid(row = 4, column = 3, pady = 1)

btnMulti = Button(innerFrame, text = "×", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.operation("multi"))
btnMulti.grid(row = 5, column = 3, pady = 1)

btndiv = Button(innerFrame, text = "÷", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.operation("divide"))
btndiv.grid(row = 6, column = 3, pady = 1)

btn0 = Button(innerFrame, text = "0", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.number_enter(0))
btn0.grid(row = 6, column = 0, pady = 1)

btnDot = Button(innerFrame, text = ".", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.number_enter("."))
btnDot.grid(row = 6, column = 1, pady = 1)

btnEqual = Button(innerFrame, text = "=", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.sum_of_total)
btnEqual.grid(row = 6, column = 2, pady = 1)

btnX1 = Button(innerFrame, text = "1/X", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.one_in)
btnX1.grid(row = 7, column = 0, pady = 1)

btnSin = Button(innerFrame, text = "sin", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.sin)
btnSin.grid(row = 7, column = 1, pady = 1)

btnCos = Button(innerFrame, text = "cos", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.cos)
btnCos.grid(row = 7, column = 2, pady = 1)

btnTan = Button(innerFrame, text = "tan", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.tan)
btnTan.grid(row = 7, column = 3, pady = 1)

btnFloor_Division = Button(innerFrame, text = "//", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.operation("floor_div"))
btnFloor_Division.grid(row = 8, column = 0, pady = 1)

btnMod = Button(innerFrame, text = "%", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.operation("mod"))
btnMod.grid(row = 8, column = 1, pady = 1)

btnExp = Button(innerFrame, text = "Exp", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = lambda:added_value.operation("exp"))
btnExp.grid(row = 8, column = 2, pady = 1)

btnEuler = Button(innerFrame, text = "e", width = 6, height = 2, font = ("arial",18,"bold"),bd = 7, bg = "gainsboro", command = added_value.euler_number)
btnEuler.grid(row = 8, column = 3, pady = 1)

#Now every buttons are in place

#Time to go back to class and write some functions!


root.mainloop()
```
