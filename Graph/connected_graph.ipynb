# Connected Graph

 Create a program which takes a graph as an input and outputs whether every node is connected or not.

## Method 1: networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

elist = [(1,1),(2,3),(2,4),(3,5),(2,3),(4,5)]

G.add_edges_from(elist)

nx.draw(G, pos=nx.circular_layout(G), node_color='r', edge_color='b')
```

```python
def connection_check(graph):
    
    check = True

    for nodea,nodeb in G.edges:
        if nodea == nodeb:
            check = False
    
    if check:
        return('Every node is connected.')
    else:
        return('Not every node is connected')
```

```python
connection_check(G)
```
